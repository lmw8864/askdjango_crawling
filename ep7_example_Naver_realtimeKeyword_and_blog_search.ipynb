## 네이버 블로그 검색

### 전체 코드

```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
from collections import OrderedDict
from itertools import count


def naver_blog_search(q, max_page=None):
    url = 'https://search.naver.com/search.naver?'
    
#     post_dict = {}
    post_dict = OrderedDict()
    
    for page in count(1):
        params = {
            'where': 'post',
            'query': q,
            'start': (page-1)*10 + 1,
        }
        
#         print(params)  # 1, 11, 21, 31, 41, 51

        ua = UserAgent()
        headers = {
                'User-Agent': ua.chrome
        }

        r = requests.get(url, params=params, headers=headers)
        html = r.text
        soup = BeautifulSoup(html, 'html.parser')

        tag_list = soup.select('.sh_blog_title')
        for tag in tag_list:
            # 멤버십 체크
            if tag['href'] in post_dict:
                return post_dict
            post_dict[tag['href']] = tag.text
    
        # max_page
        if max_page and (page >= max_page):
            break
            
    return post_dict
```

```python
result = naver_blog_search('askdjango', 2)

for key, value in result.items():
    print(value, key)
```

- 리스트나 사전을 만들어서 **멤버십 체크** 코드 추가
- 사전에서 순서를 유지하고 싶다면? collections의 OrderedDict

```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent

from collections import OrderedDict  # 순서있는 사전

def naver_blog_search(q):
    url = 'https://search.naver.com/search.naver?'
    
#     post_dict = {}
    post_dict = OrderedDict()
    
    for page in range(1, 1000):
        params = {
            'where': 'post',
            'query': q,
            'start': (page-1)*10 + 1,
        }
        
        print(params)  # 1, 11, 21, 31, 41, 51

        ua = UserAgent()
        headers = {
                'User-Agent': ua.chrome
        }

        r = requests.get(url, params=params, headers=headers)
        html = r.text
        soup = BeautifulSoup(html, 'html.parser')

        tag_list = soup.select('.sh_blog_title')
        for tag in tag_list:
            # 멤버십 체크
            if tag['href'] in post_dict:
                return post_dict
            
#             print(tag.text, tag['href'])
            post_dict[tag['href']] = tag.text  # Key로는 url을 씀.(∵ 다른 url인데 레이블은 같을 수도 있으니까)        
            
    return post_dict
```
